# **vanilla LSTM**



**VANILA LSTM IS THE LSTM MODEL WITH** 
1. SINGLE HIDDEN LAYER 
2. SINGLE LSTM UNITS
3. OUTPUT LAYER

```python
import tensorflow as tf
```

```python
import numpy as np
```

# creating a univarient data

[10,20,30,40,50,60,70,80,90]

# Turn the data into multiple input and output
# as each sample has specified time steps

10,20,30,40,50 ------- 60

20,30,40,50,60 --------70

30,40,50,60,70 --------80

```python
#creating split function
def seq_split(data, time):
    x,y = [],[]
    if time < len(data):
        for i in range(len(data)):

            x.append(data[i:time])
            y.append(data[time])
            time += 1

            if time > len(data) -1:
                break
    else:
        print("steps exceeds the splitting")
        
    return x,y
```

```python
x = [10,20,30,40,50,60,70,80,90]
new_x, new_y = seq_split(x,3)
for i in range(len(new_x)):
    print(new_x[i], new_y[i])
```

```python
new_x=np.array(new_x)
new_y=np.array(new_y)
```

# CREATING MODEL

```python
model =tf.keras.models.Sequential(
[
   
    tf.keras.layers.LSTM(50, activation = 'relu',input_shape=(3,1)),
    tf.keras.layers.Dense(1)
]
                                )
```

```python
model.compile(optimizer='adam',
             loss='mse',
             metrics=['mae'])
```

```python
x = new_x.reshape((6,3,1))
y = new_y
```

```python
model.fit(x,y,epochs=200)
```

```python
x_input = np.array([100, 110, 120])
x_input = x_input.reshape((1, 3,1))
model.predict(x_input)
```
